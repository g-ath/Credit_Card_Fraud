```python
def Unemployment(code):
    code['DATE'] = code['DATE'].astype('datetime64[ns]')
    code['Year'] = [x.year for x in code['DATE']]
    code['Unemployment'] = code.iloc[:,1].apply(lambda x: x)
    code = code.drop(code.columns[1], axis=1)
    code = code.drop(['DATE'], axis = 1)
    return code
```

```python
def merge_df(users = users, cards = cards, trans = trans):
    users['User'] = users.index
    first_column = users.pop('User')
    users.insert(0, 'User', first_column)
    cards['Card'] = cards['CARD INDEX']
    cards = cards.drop('CARD INDEX', axis = 1)
    df_merge = pd.merge(trans, users, how = 'inner', on = 'User')
    df_merge = pd.merge(df_merge, cards, how = 'inner', on = ['User', 'Card'])
    df_merge = df_merge.sort_values(by = 'User', axis=0, ascending=True)
    df_merge = df_merge.reset_index()
    column_to_reorder = df_merge.pop('Is Fraud?')
    df_merge.insert(len(df_merge.columns), 'Is Fraud?', column_to_reorder)
    return df_merge
```

```python
def strat_split(df_merge = df_merge, test_size = 0.25):
    X = df_merge.iloc[:,:-1]
    y = df_merge.iloc[:,-1]
    X_strat_train, X_final_test, y_strat_train, y_final_test = train_test_split(X, y, stratify=y, test_size=test_size)
    strat_train = pd.concat([X_strat_train, y_strat_train], axis = 1, join="inner")
    strat_test = pd.concat([X_final_test, y_final_test], axis = 1, join="inner")
    strat_train = strat_train.sort_values(by = 'User', axis=0, ascending=True)
    strat_test = strat_test.sort_values(by = 'User', axis=0, ascending=True)
    return strat_train, strat_test
```
